```python
import numpy as np
```

```python
array=np.arange(1,9)
array
```

```python
#indexing of array
array[1]
```

```python
#now we are seeing how we can slice or index matrice
array2=np.arange(1,21).reshape(5,4)
array2
```

```python
#indexing of columns is easy
array2[2]
```

```python
#if we have to slice the elements between the matrice
array2[2,3]
```

```python
#if we have to find the elements like 14 15 16 18 19 20 so
array2[3:,1:]
```

```python
# if we have to find rows
array2[0:,:0]
```
